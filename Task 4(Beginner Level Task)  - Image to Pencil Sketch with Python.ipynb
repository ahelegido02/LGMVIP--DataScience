```python
import cv2
import sys
import matplotlib.pyplot as plt

def convert_image_to_pencil_sketch(image_path, output_path):
    # Load the image
    image = cv2.imread(image_path)

    if image is None:
        print("Error: Image not found.")
        sys.exit()

    # Convert the image to grayscale
    gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

    # Invert the grayscale image
    inverted_gray_image = 255 - gray_image

    # Apply Gaussian Blur to the inverted image
    blurred_image = cv2.GaussianBlur(inverted_gray_image, (21, 21), 0)

    # Invert the blurred image
    inverted_blurred_image = 255 - blurred_image

    # Create the pencil sketch by mixing the grayscale image with the inverted blurred image
    pencil_sketch = cv2.divide(gray_image, inverted_blurred_image, scale=256.0)

    # Save the pencil sketch to the output file
    cv2.imwrite(output_path, pencil_sketch)

    # Convert BGR to RGB for displaying with matplotlib
    image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    pencil_sketch_rgb = cv2.cvtColor(pencil_sketch, cv2.COLOR_GRAY2RGB)

    # Display the original image and the pencil sketch side by side
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(20, 10), gridspec_kw={'wspace': -0.26})
    ax1.imshow(image_rgb)
    ax1.set_title('Original Image')
    ax1.axis('off')
    ax2.imshow(pencil_sketch_rgb)
    ax2.set_title('Pencil Sketch')
    ax2.axis('off')
    plt.show()

if __name__ == "__main__":
    image_path = "Dog.png"                  # Replace with your image path
    output_path = "pencil_sketch.jpg"       # Replace with your desired output path
    convert_image_to_pencil_sketch(image_path, output_path)
```

**<u><font size="+0">SUMMARY:</font></u>**

This Python script is designed to convert an input image into a pencil sketch. The script uses the OpenCV library for image processing and the matplotlib library for displaying the results. Here's a summary of the script's functionality:

1. Load an image from the specified path.
2. Convert the image to grayscale to simplify the processing.
3. Invert the grayscale image to enhance the details.
4. Apply a Gaussian blur to the inverted image to smooth the edges.
5. Invert the blurred image back to the original orientation.
6. Combine the grayscale image with the inverted blurred image to create a pencil sketch effect.
7. Save the resulting pencil sketch to a specified output file.
8. Display the original image and the pencil sketch side by side using matplotlib.
